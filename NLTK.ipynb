```python
def lexical_diversity(text):
    # the ratio of unique words to total words of a text
    return len(set(text)) / len(text)

def percentage(count, total):
    # percentage a word or character makes up total text
    return 100 * count / total
```

```python
lexical_diversity(text3)
```

```python
percentage(text4.count('a'), len(text4))
```

```python
from nltk.corpus import gutenberg

# print average number of characters per word, average number of words per sentence, and average repetition of words

for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    
    print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
```

```python
# Inaugural addresses of US presidents

from nltk.corpus import inaugural
```

```python
# Frequency of the words 'citizen' and 'america' over time

cfd = nltk.ConditionalFreqDist((target, fileid[:4])
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    for target in ['america', 'citizen']
    if w.lower().startswith(target))
    
cfd.plot()
```
